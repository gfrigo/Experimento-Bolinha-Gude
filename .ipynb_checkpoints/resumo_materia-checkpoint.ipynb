## Tipos de Erros

#### Erros Sistemáticos

`DEFINIÇÃO`:
São causados por fontes identificáveis, podem ser eliminados ou compensados e fazem com que as medidas estejam acima ou abaixo prejudicando a accuracy.

`CAUSAS:`
* Instrumento: erros causados em medidas de intervalo de tempo feitas com um relógio que atrasa;
* Método de pbservação: medir o instante de ocorrência de um relâmpago pelo ruído do trovão associado;
* Efeitos ambientais: uma medida de frequência da luz emitida
por um laser, dependendente da temperatura ambiente;
* Simplificações do modelo teórico: Evitar conceitos físicos em análises, ex: desconsiderar o efeito de resistência do ar ao analisar a gravidade.

`TRATAMENTO:`
* Eliminar o maior número possível de erros sistemáticos.

#### Erros Aleatórios

`DEFINIÇÃO`:
São flutuações, para cima ou para baixo, de modo que, metade das medidas, de mesma grandeza, estejam desviadas para cima e a outra metade desviadas para baixo.

`FONTES:`
Nem sempre são identificáveis, então podem ser:
* Método de observação: erro devido ao julgamento feito
pelo observador ao fazer uma leitura abaixo da menor divisão
de uma escala;
* Flutuações Ambientais: mudanças não previsíveis de situações ambientais.

`TRATAMENTO:` Podem ser tratados quantitativamente através de métodos estatísticos.
